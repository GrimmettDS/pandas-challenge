```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# File to Load
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"
```

```python
# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
# Combine the data into a single dataset.
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

### District Summary

```python
# Calculate the total number of schools
# Calculate the total number of students
# Calculate the total budget
# Calculate the average math score
# Calculate the average reading score
# Calculate the percentage of students with a passing math score (70 or greater)
# Calculate the percentage of students with a passing reading score (70 or greater)
# Calculate the percentage of students who passed math and reading (% Overall Passing)
# Create a dataframe to hold the above results
# Optional: give the displayed data cleaner formatting
```

```python
schools = school_data_complete['school_name'].unique()
school_count = len(schools)
school_count
```

```python
students = school_data_complete['student_name'].count()
students
```

```python
budget = school_data_complete['budget'].unique()
budget_sum = budget.sum()
budget_sum
```

```python
avg_math_score = school_data_complete['math_score'].sum() / students
avg_math_score
```

```python
avg_reading_score = school_data_complete['reading_score'].sum() / students
avg_reading_score
```

```python
pass_math = school_data_complete[school_data_complete["math_score"]>= 70]
pass_math_count = pass_math["student_name"].count()
pass_math_pct = (pass_math_count / students) * 100
pass_math_pct
```

```python
pass_reading = school_data_complete[school_data_complete["reading_score"]>= 70]
pass_reading_count = pass_reading["student_name"].count()
pass_reading_pct = (pass_reading_count / students) * 100
pass_reading_pct
```

```python
pass_both = school_data_complete.loc[(school_data_complete['math_score'] >= 70) & 
                                     (school_data_complete['reading_score'] >= 70), :]
pass_both = pass_both["Student ID"].count()
#pass_both
#len(pass_both) - without this the number is 25528
pass_both_pct = (pass_both / students) * 100
pass_both_pct
```

```python
# Creating a summary DataFrame using above values
summary_df = pd.DataFrame({"Total Schools":[school_count],
                          "Total Students":[students],
                          "Total Budget": [budget_sum],
                          "Average Math Score": [avg_math_score],
                          "Average Reading Score": [avg_reading_score],
                          "% Passing Math": [pass_math_pct],
                          "% Passing Reading": [pass_reading_pct],
                          "% Overall Passing": [pass_both_pct],
                          })
summary_df["Total Students"] = summary_df["Total Students"].map('{:,}'.format)
summary_df["Total Budget"] = summary_df["Total Budget"].map('${:,.2f}'.format)
summary_df["Average Math Score"] = summary_df["Average Math Score"].map('{:.3f}'.format)
summary_df["Average Reading Score"] = summary_df["Average Reading Score"].map('{:.3f}'.format)
summary_df["% Passing Math"] = summary_df["% Passing Math"].map('{:.2f}%'.format)
summary_df["% Passing Reading"] = summary_df["% Passing Reading"].map('{:.2f}%'.format)
summary_df["% Overall Passing"] = summary_df["% Overall Passing"].map('{:.2f}%'.format)

summary_df
```

### School Summary

```python
# Create an overview table that summarizes key metrics about each school including:
    # School Name
    # School Type
    # Total Students
    # Total School Budget
    # Per Student Budget
    # Average Math Score
    # Average Readiing Score
    # % Passing Math
    # % Passing Reading
    # % Overall Passing (The percentage of students that passed math and reading.)
    
# Create a dataframe to hold the results.
```

```python
# Groups by school
index_by_school = school_data_complete.set_index('school_name').groupby(['school_name'])
```

```python
# School types
sch_types = school_data.set_index('school_name')['type']
```

```python
# Total students by school with data check
stu_per_sch = index_by_school['Student ID'].count()
stu_per_sch
```

```python
# School budget with data check
sch_budget = school_data.set_index('school_name')['budget']
sch_budget
```

```python
# Per student budget with data check
stu_budget = school_data.set_index('school_name')['budget']/school_data.set_index('school_name')['size']
stu_budget
```

```python
# Avg scores by school with data check
avg_math = index_by_school['math_score'].mean()
avg_reading = index_by_school['reading_score'].mean()

#avg_math
avg_reading
```

```python
# Percent passing scores
pass_math = school_data_complete[school_data_complete['math_score'] >= 70].groupby('school_name')['Student ID'].count()/stu_per_sch 
pass_reading = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('school_name')['Student ID'].count()/stu_per_sch 
pass_both = school_data_complete[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)].groupby('school_name')['Student ID'].count()/stu_per_sch

pass_math = pass_math * 100
pass_reading = pass_reading * 100
pass_both = pass_both * 100
```

```python
# Creating dataframe for school results - includes format changes

school_summary = pd.DataFrame({"School Type": sch_types,
                            "Total Students": stu_per_sch,
                            "Total School Budget": sch_budget,
                            "Per Student Budget": stu_budget,
                            "Average Math Score": avg_math,
                            "Average Reading Score": avg_reading,
                            "% Passing Math": pass_math,
                            "% Passing Reading": pass_reading,
                            "Overall Passing Rate": pass_both})

school_summary["Total Students"] = school_summary["Total Students"].map('{:,}'.format)
school_summary["Total School Budget"] = school_summary["Total School Budget"].map('${:,.2f}'.format)
school_summary["Per Student Budget"] = school_summary["Per Student Budget"].map('${:,.2f}'.format)
school_summary["Average Math Score"] = school_summary["Average Math Score"].map('{:.3f}'.format)
school_summary["Average Reading Score"] = school_summary["Average Reading Score"].map('{:.3f}'.format)
school_summary["% Passing Math"] = school_summary["% Passing Math"].map('{:.2f}%'.format)
school_summary["% Passing Reading"] = school_summary["% Passing Reading"].map('{:.2f}%'.format)
school_summary["Overall Passing Rate"] = school_summary["Overall Passing Rate"].map('{:.2f}%'.format)

school_summary
```
